```python
new_mask = np.zeros((64, 64), dtype=np.uint8)
```

```python
for x in range(new_label.shape[0]):
    for y in range(new_label.shape[1]):
        for c in range(new_label.shape[2]):
            if new_label[x,y,c] == 128:
                new_mask[x,y] = c+1
```

```python
imshow(new_mask)
```

```python
imsave('new_mask.png', new_mask)
```
